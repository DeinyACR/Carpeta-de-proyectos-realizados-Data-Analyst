* Cambiar los nombres a las columnas de modo que queden tipo 'snake_case'
* En la columna name:
    * Cambiar los nombres a mayusculas y quitar espacios no deseados y signos de pausa.
    * Rellenar los casos nulos con 'NO DEFINED' 
* Columna platform: 
    * Agrupar las plataformas por tipolog√≠a y crear una columna nueva. 
* Columna year_of_release:
    * Quitar valores perdidios.
* Columna genre: 
    * Rellenar los datos nulos con 'No defined'
* Columna user_score:
    * Sustituir 'tbd' por un valor numerico perdido.
    * Cambiar los datos a tipo de datos float.
* Columna 'critic_Score:
    * dividir los datos entre 10 para tener las mismas magnitudes que 'user_score'
* Columna rating:
    * Cambiar las opciones existentes por valores mas explicitos para el publico. 
* Generar una nueva columna 'total_sales' 
* Revisar y eliminal doplicados. 

### Desarrollo 

Nombres de columnas

```python
df_games.columns = [x.lower() for x in df_games.columns]
df_games.info()
```

Columna 'names'

```python
# Cambiar a mayusculas
df_games['name'] = df_games['name'].str.upper()
```

```python
# Quitar espacios no deseados y signos de pausa
df_games['name'] = df_games['name'].str.strip()
df_games['name'] = df_games['name'].str.replace(':','').str.replace(';','').str.replace(',','').str.replace('.','')
```

```python
df_games.sample(15)
```

```python
df_games[df_games['name'].isna()]
```

```python
# Rellenar valores nulos
df_games['name'] = df_games['name'].fillna('NOT DEFINED')
```

Columna platform

```python
# Crear una funcion a aplicar
def gruoplat (x):
    if x in ['PS', 'PS2', 'PSP', 'PS3', 'PSV', 'PS4', 'PS5']:
        return 'PlayStation'
    elif x in ['X360', 'XB', 'XOne']:
        return 'XBox'
    elif x in ['3DS', 'DS', 'GB', 'GBA', 'GC', 'N64', 'NES', 'SNES', 'Wii', 'WiiU']:
        return 'Nintendo'
    elif x in ['PC', 'PCFX']:
        return 'PC'
    else:
        return 'Other'
```

```python
# Aplicar la funcion para crear la nueva columna 
df_games['platform_group'] = df_games['platform'].apply(gruoplat)
df_games.sample(15)
```

```python
df_games['platform_group'].value_counts()
```

Columna user_score

```python
# Cambiar tbd por un valor numerico perdido
df_games.loc[df_games['user_score'] == 'tbd', 'user_score'] = np.nan
```

```python
# Cambiar el tipo de datos
df_games['user_score'] = df_games['user_score'].astype(float)
```

```python
df_games['user_score'].describe()
```

Columna critic_score 

```python
# Igualar la magnitud con 'user_score'
df_games['critic_score'] = df_games['critic_score']/10
```

```python
df_games.sample(15)
```

Columna 'rating'

```python
# Definir una funcion a plaicar 
def concept(x):
    if x == 'E' or x == 'K-A':
        return 'Everyone'
    elif x == 'T':
        return 'Teens'
    elif x == 'M':
        return 'Mature'
    elif x == 'E10+':
        return 'Everyone 10+'
    elif x == 'EC':
        return 'Early Childhood'
    elif x == 'RP':
        return 'Rating Pending'
    elif x == 'AO':
        return 'Adults Only'
    else:
        return 'Rating Pending'
```

```python
df_games['rating'] = df_games['rating'].apply(concept)
```

```python
df_games['rating'].value_counts()
```

Columna total_sales

```python
df_games['total_sales'] = df_games[['na_sales', 'eu_sales', 'other_sales']].sum(axis = 1)
```

```python
df_games.sample(15)
```
