```python
# Histograma de las ventas de Nintendo
sales_nintendo.plot(kind = 'hist', 
                    bins = 20, 
                    title = 'Dsitribucion de las ventas de Nintendo', 
                    grid = True, 
                    xlabel = 'Valor de las ventas', 
                    ylabel = 'Frecuencia', 
                    color = 'red')
print(f'La mediana de las ventas anuales de Nintendo es: {sales_nintendo.median().round()}')
print(f'La media de las ventas anuales de Nintendo es: {sales_nintendo.mean().round()}')
```

Podemos ver que la mayoria de las ventas de Nintendo se entcuentra entre los valores 0 y 100, concluyendo que sus ventas no son tan altas como las de otras empresas lideres.

## Prueba de Hipótesis

### Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas

Hipotesis nula:Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son iguales.

Hipotesis alternativa:Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son distintas.

Nivel de significancia de estadística crítica: 5% 

Se utilizara la función scipy.stats.ttest_ind(array1, array2, equal_var) en las dos hipotesis ya que las muestras se toman de poblaciones con parámetros similares. En este caso equal_var se establece con True 

```python
hip_1 = df_final.query("platform == 'PC' or platform == 'XOne'")
hip_1 = hip_1.pivot_table(index = 'name',
                  columns = 'platform',
                  values = 'user_score')
hip_1 = hip_1.dropna()
```

```python
alpha = 0.05 

results = st.ttest_ind(hip_1['PC'], hip_1['XOne']) 

print('valor p:', results.pvalue)

if (results.pvalue < alpha):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

Se obtuvo un valor de 0.021 en p, ya que este valor es menor al nivel de significancia se rechaza la hipotesisi nula, lo que sugiere que los promedios de las calificaciones de PC y XBox son diferentes. 

#### Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.

Hipotesis nula:Las calificaciones promedio de los usuarios para los juego de acción y deportes son iguales.

Hipotesis alternativa:Las calificaciones promedio de los usuarios para los juego de acción y deportes son diferentes.

Nivel de significancia de estadística crítica: 5% 

Se utilizara la función scipy.stats.ttest_ind(array1, array2, equal_var) en las dos hipotesis ya que las muestras se toman de poblaciones con parámetros similares. En este caso equal_var se establece con True 

```python
hip_2 = df_final.query("genre == 'Action'")
hip_2 = hip_2.pivot_table(index = 'name',
                          columns = 'genre',
                          values = 'user_score')
hip_2 = hip_2.dropna()
hip_2
```

```python
hip_2_1 = df_final.query("genre == 'Sports'")
hip_2_1 = hip_2_1.pivot_table(index = 'name',
                            columns = 'genre',
                            values = 'user_score')
hip_2_1 = hip_2_1.dropna()
hip_2_1
```

```python
alpha = 0.05 

results = st.ttest_ind(hip_2, hip_2_1) 

print('valor p:', results.pvalue)

if (results.pvalue < alpha):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

Se obtuvo un valor de 1.45597414e-11 en p, ya que este valor es menor al nivel de significancia se rechaza la hipotesisi nula, lo que sugiere que los promedios de las calificaciones de los juegos de acción y deportes son diferentes.
